```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
```

## chrome driver

```python
os.environ['PATH']+= r"D:/Resources/Project/webscrap/chromedriver_win32"
driver = webdriver.Chrome()
```

## implementation

```python
# foods nutritions attributes list
Name_l=[]
Serving_Size_l=[]
Calories_l=[]
Total_Fat_l=[]
Cholesterol_l=[]
Sodium_l=[]
Potassium_l=[]
Total_Carbohydrates_l=[]
Protein_l=[]
Vitamin_A_l=[]
Vitamin_C_l=[]
Calcium_l=[]
Iron_l=[]
```

```python
# function to get the single food nutrition data from the website

def food_details():

    food_name =driver.find_element('xpath',"//div[contains(@class, 'name')]")
    Name_l.append(food_name.text)
    
    food_size = driver.find_element('xpath',"//span[contains(@itemProp, 'servingSize')]")
    Serving_Size_l.append(food_size.text.splitlines()[0])
    
    calories = driver.find_element('xpath',"//span[contains(@itemProp, 'calories')]")
    Calories_l.append(calories.text.splitlines()[0])
    
    total_fat = driver.find_element('xpath',"//span[contains(@itemProp, 'fatContent')]")
    Total_Fat_l.append(total_fat.text.splitlines()[0])
    
    Cholesterol = driver.find_element('xpath',"//span[contains(@itemProp, 'cholesterolContent')]")
    Cholesterol_l.append(Cholesterol.text.splitlines()[0])
    
    Sodium = driver.find_element('xpath',"//span[contains(@itemProp, 'sodiumContent')]")
    Sodium_l.append(Sodium.text.splitlines()[0])
    
    Potassium  = driver.find_element('xpath',"//span[contains(@itemProp, 'potassiumContent')]")
    Potassium_l.append(Potassium .text.splitlines()[0])
    print(Potassium .text.splitlines()[0],'\n')
    
    Total_Carbohydrates  = driver.find_element('xpath',"//span[contains(@itemProp, 'carbohydrateContent')]")
    Total_Carbohydrates_l.append(Total_Carbohydrates.text.splitlines()[0])
  
    Protein = driver.find_element('xpath',"//span[contains(@itemProp, 'proteinContent')]")
    Protein_l.append(Protein.text.splitlines()[0])
    
    Vitamin_A_per = driver.find_element(By.CSS_SELECTOR, 'div.vitaminA>div.dv')
    Vitamin_A_l.append(Vitamin_A_per.text)
    
    Vitamin_C_per = driver.find_element(By.CSS_SELECTOR, 'div.vitaminC>div.dv')
    Vitamin_C_l.append(Vitamin_C_per.text)
    
    calcium_per = driver.find_element(By.CSS_SELECTOR, 'div.calcium>div.dv')
    Calcium_l.append(calcium_per.text)
    
    iron_per = driver.find_element(By.CSS_SELECTOR, 'div.iron>div.dv')
    Iron_l.append(iron_per.text)
    
    
```

```python
# read the food list and opens website,then calls food_details function
with open('1_2_3_food.txt') as f:
    lines = f.readlines()
    i=0
    for foods in lines:
        
        #search foods
        query = foods     
        link = 'https://www.nutritionix.com/search?q='+query
        try:
            search = driver.get(link) 
            driver.implicitly_wait(5)
            food = driver.find_element(By.CSS_SELECTOR,'ul.list-search-results>li.ng-scope>a').get_attribute('href')
            #click food
            # try:
            #     food = driver.find_element("xpath","/html/body/div[2]/div/div[3]/div/div[1]/ul/li[1]/a").get_attribute('href')
            # except:
            #     food = driver.find_element("xpath","/html/body/div[2]/div/div[3]/div/div[1]/ul/li/a").get_attribute('href')
            
            driver.implicitly_wait(5)                                
            search = driver.get(food)  
            driver.implicitly_wait(5)
            i+=1
            print(i, 'done')
            # food details fill
            food_details()
        except:
            i+=1
            print(i,foods,'not done')
```

```python
# store data into dataframe
import pandas as pd

df = pd.DataFrame()
df['Name']=Name_l
df['Serving_Size']=Serving_Size_l
df['Calories']=Calories_l
df['Total_Fat']=Total_Fat_l
df['Cholesterol']=Cholesterol_l
df['Sodium']=Sodium_l
df['Potassium']=Potassium_l
df['Total_Carbohydrates']=Total_Carbohydrates_l
df['Protein']=Protein_l
df['Vitamin_A']=Vitamin_A_l
df['Vitamin_C']=Vitamin_C_l
df['Calcium']=Calcium_l
df['Iron']=Iron_l
df.head()
```

```python
# save as csv file
df.to_csv('food_nutrition_1.csv')
```
